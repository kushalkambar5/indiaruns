```python
'candidate_id', 'name', 'location','country', 
rows_list = []


for candidate in candidates:
    profile = candidate.get('profile', {})
    career_history = candidate.get('career_history', [])
    education = candidate.get('education', [])
    skills = candidate.get('skills', [])
    certifications = candidate.get('certifications', {})
    languages = candidate.get('languages', {})
    redrob_signals = candidate.get('redrob_signals', {})

    isNoidaorPune = 0
    if 'noida' in profile.get('location').lower() or 'pune' in profile.get('location').lower():
        isNoidaorPune=1

    isIndia=0
    if 'india' in profile.get('country').lower():
        isIndia = 1
    
    def extract_grade(grade_str):
        value = float(grade_str.split()[0].replace('%', ''))
        if '%' in grade_str:
            value /= 10.0
        return value

    if education:
        tiers = [
            float(ed['tier'].split('_')[1])
            for ed in education
            if ed.get('tier')
        ]
        college_tier = min(tiers) if tiers else np.nan
        grades = [
            extract_grade(ed['grade'])
            for ed in education
            if ed.get('grade')
        ]
        grade = max(grades) if grades else np.nan
    else:
        college_tier = np.nan
        grade = np.nan

    age = 19
    if len(education)>1:
        age = age + education[0]['end_year'] - education[0]['start_year'] + education[1]['end_year'] - education[1]['start_year'] + 2
    else:
        age = age + education[0]['end_year'] - education[0]['start_year'] + 1

    


    preffered_degrees = ['cs', 'it', 'ai', 'computer', 'information', 'artificial', 'machine', 'learning', 'mnc', 'computing', 'data']
    is_cs_related_branch = 0
    if education:
        records_to_check = education[:2]
        fields = [req.get('field_of_study', '').lower() for req in records_to_check]
        combined_fields_text = " ".join(fields)
        if any(degree in combined_fields_text for degree in preffered_degrees):
            is_cs_related_branch = 1


    import re

    size_str = candidate['profile']['current_company_size']
    match = re.search(r'\d+', size_str)

    if match:
        number = int(match.group())

    is_startup=0
    if float(number) < 100:
        is_startup=1

    consultancy_companies = ['tcs', 'infosys', 'wipro', 'accenture', 'cognizant', 'capgemini', 'mindtree']
    consultancy_industries = ['it', 'it solutions', 'service', 'services', 'consultancy']
    current_company = candidate.get('profile', {}).get('current_company', '').lower()
    current_industry = candidate.get('profile', {}).get('current_industry', '').lower()
    is_consultancy = 0
    if any(company in current_company for company in consultancy_companies):
        is_consultancy = 1
    elif any(industry in current_industry for industry in consultancy_industries):
        is_consultancy = 1


    last_active_date = redrob_signals['last_active_date']
    reference_date = pd.Timestamp("2026-06-17")
    target_date = pd.Timestamp(last_active_date)
    last_active_days_ago = (reference_date - target_date).days

    
    # Language Score

    WEIGHTS = {
        'english': 0.6,
        'hindi': 0.25,
        'default': 0.1
    }

    PROFICIENCIES = {
        'native': 1.0,
        'professional': 0.75,
        'conversational': 0.5
    }

    languages_score = 0.0

    for entry in languages:
        lang_name = entry['language'].lower()
        prof_level = entry['proficiency'].lower()
    
        base_weight = WEIGHTS.get(lang_name, WEIGHTS['default'])
    
        multiplier = PROFICIENCIES.get(prof_level, 0.5)
    
        languages_score += base_weight * multiplier



    row_data = {
        'candidate_id': candidate.get('candidate_id'),
        'name': profile.get('anonymized_name'),
        'location': profile.get('location'),
        'country': profile.get('country'),
        'years_of_experience': profile['years_of_experience'],
        'isNoidaorPune': isNoidaorPune,
        'isIndia': isIndia,
        'college_tier': college_tier,
        'grade': grade,
        'no_of_degrees': len(education),
        'age':age,
        'is_cs_related_branch':is_cs_related_branch,
        'is_startup': is_startup,
        'is_consultancy': is_consultancy,
        'profile_completeness_score': redrob_signals['profile_completeness_score'],
        'last_active_days_ago': last_active_days_ago,
        'open_to_work_flag': 0 if redrob_signals['open_to_work_flag']==False else 1,
        'profile_views_received_30d': redrob_signals['profile_views_received_30d'],
        'applications_submitted_30d': redrob_signals['applications_submitted_30d'],
        'recruiter_response_rate': redrob_signals['recruiter_response_rate'],
        'avg_response_time_hours': redrob_signals['avg_response_time_hours'],
        'connection_count': redrob_signals['connection_count'],
        'endorsements_received': redrob_signals['endorsements_received'],
        'notice_period_days': redrob_signals['notice_period_days'],
        'min_lpa': redrob_signals['expected_salary_range_inr_lpa']['min'],
        'preferred_work_mode': redrob_signals['preferred_work_mode'],
        'willing_to_relocate': 0 if redrob_signals['willing_to_relocate']==False else 1,
        'github_activity_score': redrob_signals['github_activity_score'],
        'search_appearance_30d': redrob_signals['search_appearance_30d'],
        'saved_by_recruiters_30d': redrob_signals['saved_by_recruiters_30d'],
        'interview_completion_rate': redrob_signals['interview_completion_rate'],
        'offer_acceptance_rate': redrob_signals['offer_acceptance_rate'],
        'verified_email': 0 if redrob_signals['verified_email']==False else 1,
        'verified_phone': 0 if redrob_signals['verified_phone']==False else 1,
        'linkedin_connected': 0 if redrob_signals['linkedin_connected']==False else 1,
        'languages_score': languages_score
    }
    rows_list.append(row_data)




df = pd.DataFrame(rows_list)    
```
